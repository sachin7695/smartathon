```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

```python
files = os.listdir("imagesss/images/")
my_dict = {}
my_dict["bboxes"] = []
for index, row in df.iterrows():
    keys = ["class", "filename", "xmin", "ymin", "xmax", "ymax"]
    values = [row["class"], row["image_path"], row["xmin"], row["ymin"], row["xmax"], row['ymax']]
    bbox = {}
    bbox.update((key, value) for key, value in zip(keys, values))
    my_dict["bboxes"].append(bbox)
```

```python
my_dict["bboxes"][0]["filename"]
len(my_dict["bboxes"])
my_dict
```

```python
my_final_dict = {}
for box in my_dict["bboxes"]:
    if box["filename"] not in my_final_dict:
        my_final_dict[box["filename"]] = []
    my_final_dict[box["filename"]].append({
        "class": box["class"],
        "xc": ((2*box["xmin"]+2*box["xmax"]) / 2)/1920,
        "yc": ((2*box["ymin"] + 2*box["ymax"]) / 2) /1080,
        "width": (2*box["xmax"] - 2*box["xmin"])/1920,
        "height": (2*box["ymax"] - 2*box["ymin"])/1080
    })
```
